```python
import pandas as pd
from openpyxl import load_workbook

# === STEP 1: Load file mentah ===
df = pd.read_csv("diabetes_dataset.csv")  # ganti nama file jika beda

# === STEP 2: Mapping kategori ke angka ===
mapping = {
    'Yes': 1, 'No': 0,
    'Positive': 1, 'Negative': 0,
    'High': 2, 'Moderate': 1, 'Low': 0,
    'Healthy': 1, 'Unhealthy': 0,
    'Normal': 1, 'Abnormal': 0,
    'Present': 1, 'Absent': 0
}
df.replace(mapping, inplace=True)
df = df.infer_objects(copy=False)


# === STEP 3: Mapping nilai khusus "Urine Test" ===
urine_map = {
    'Ketones Present': 1,
    'Glucose Present': 2,
    'Protein Present': 3
}
if 'Urine Test' in df.columns:
    df['Urine Test'] = df['Urine Test'].map(urine_map)

# === STEP 4: Lowercase + trim kolom "Target"
if 'Target' in df.columns:
    df['Target'] = df['Target'].str.strip().str.lower()

# === STEP 5: Isi missing value
df.fillna(df.median(numeric_only=True), inplace=True)

# === STEP 6: Simpan ke CSV hasil cleaned
df.to_csv("diabetes_cleaned_final.csv", index=False)

# === STEP 7: Simpan ke Excel
excel_file = "diabetes_cleaned_final.xlsx"
df.to_excel(excel_file, index=False)

# === STEP 8: Rapikan file Excel (autofit, freeze, filter)
wb = load_workbook(excel_file)
ws = wb.active

# Autofit lebar kolom
for column_cells in ws.columns:
    max_length = 0
    col = column_cells[0].column_letter
    for cell in column_cells:
        try:
            if cell.value:
                max_length = max(max_length, len(str(cell.value)))
        except:
            pass
    ws.column_dimensions[col].width = max_length + 2

# Freeze header dan aktifkan filter
ws.freeze_panes = "A2"
ws.auto_filter.ref = ws.dimensions

# Simpan akhir
wb.save(excel_file)

print("✅ Beres! Data sudah bersih dan disimpan ke:")
print("📄 CSV  ➤ diabetes_cleaned_final.csv")
print("📊 Excel ➤ diabetes_cleaned_final.xlsx")
```

```python
# ==============================================
# 📦 Import Library
# ==============================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import joblib

# ==============================================
# 📂 Load Dataset Cleaned
# ==============================================
df = pd.read_csv('diabetes_cleaned_final.csv')
print("📊 Dataset (5 baris awal):")
print(df.head())

# ==============================================
# 🔍 Cek Info Dataset
# ==============================================
print("\n📋 Info dataset:")
print(df.info())

# ==============================================
# 🎯 Pisahkan Fitur (X) dan Target (y)
# ==============================================
X = df.drop(columns=['Target'])  # asumsi 'Target' adalah kolom diagnosis
y = df['Target']

# Encode label target ke angka
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# ==============================================
# ✂️ Split Data: Training dan Testing
# ==============================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

# ==============================================
# 🤖 Training Model Decision Tree
# ==============================================
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# ==============================================
# 📈 Evaluasi Model
# ==============================================
y_pred = model.predict(X_test)
print("\n🎯 Akurasi Model:", accuracy_score(y_test, y_pred))
print("\n📑 Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# ==============================================
# 🌳 Visualisasi Decision Tree
# ==============================================
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=X.columns, class_names=le.classes_, filled=True)
plt.title("Decision Tree for Diabetes Diagnosis", fontsize=16)
plt.show()

# ==============================================
# 📜 Ekstrak Rules (Sistem Pakar)
# ==============================================
print("\n📚 Aturan dari Decision Tree:")
rules = export_text(model, feature_names=list(X.columns))
print(rules)

# ==============================================
# 💾 Simpan Model dan Encoder
# ==============================================
joblib.dump(model, 'diabetes_model.pkl')
joblib.dump(le, 'label_encoder.pkl')

print("\n✅ Model dan encoder berhasil disimpan!")
```

```python
# ==============================================
# 📦 Import Library
# ==============================================
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import joblib

# ==============================================
# 📂 Load Dataset Cleaned
# ==============================================
df = pd.read_csv('diabetes_cleaned_final.csv')

# ==============================================
# 🔤 Encode semua fitur kategorik
# ==============================================
X = df.drop(columns=['Target'])
y = df['Target']

le_dict = {}
for col in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    le_dict[col] = le

le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# ==============================================
# 🤖 Inisialisasi Model
# ==============================================
model = DecisionTreeClassifier(random_state=42)

# ==============================================
# 🔁 Cross-Validation (Stratified K-Fold)
# ==============================================
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X, y_encoded, cv=cv, scoring='accuracy')

print("📊 Akurasi per Fold:", cv_scores)
print("📈 Rata-rata Akurasi:", np.mean(cv_scores))

# ==============================================
# 📊 Train Final Model dengan Seluruh Data
# ==============================================
model.fit(X, y_encoded)

# ==============================================
# 🌳 Visualisasi Decision Tree
# ==============================================
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=X.columns, class_names=le_target.classes_, filled=True)
plt.title("Decision Tree Final (Full Data)")
plt.show()

# ==============================================
# 📜 Ekstrak Aturan Sistem Pakar
# ==============================================
print("\n📚 Aturan Decision Tree:")
rules = export_text(model, feature_names=list(X.columns))
print(rules)

# ==============================================
# 💾 Simpan Model dan Encoder
# ==============================================
joblib.dump(model, 'diabetes_model_cv.pkl')
joblib.dump(le_target, 'label_encoder_cv.pkl')
joblib.dump(le_dict, 'feature_encoders_cv.pkl')

print("\n✅ Model dan encoder hasil CV berhasil disimpan!")
```

```python
# ==============================================
# 📦 Import Library
# ==============================================
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import joblib
import numpy as np

# ==============================================
# 📂 Load Dataset Cleaned
# ==============================================
df = pd.read_csv('diabetes_cleaned_final.csv')
print("📊 Dataset (5 baris awal):")
print(df.head())

# ==============================================
# 🔤 Encode fitur kategorikal
# ==============================================
X = df.drop(columns=['Target'])  # fitur
y = df['Target']                 # target

le_dict = {}
for col in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    le_dict[col] = le

# Encode target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# ==============================================
# 🎯 Cross Validation
# ==============================================
model_cv = DecisionTreeClassifier(random_state=42)
cv_scores = cross_val_score(model_cv, X, y_encoded, cv=5)
print("\n📊 Akurasi per Fold:", np.round(cv_scores, 5))
print("📈 Rata-rata Akurasi:", np.mean(cv_scores))

# ==============================================
# ✂️ Split Data: Train dan Test
# ==============================================
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# ==============================================
# 🤖 Train Final Model
# ==============================================
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# ==============================================
# 📈 Evaluasi
# ==============================================
y_pred = model.predict(X_test)
print("\n🎯 Akurasi Model (Final Test):", accuracy_score(y_test, y_pred))
print("\n📑 Classification Report:")
print(classification_report(y_test, y_pred, target_names=le_target.classes_))

# ==============================================
# 🔍 Confidence Score
# ==============================================
y_proba = model.predict_proba(X_test)
print("\n🔎 Prediksi dan Confidence Score (5 sampel pertama):")
for i in range(5):
    label = le_target.inverse_transform([y_pred[i]])[0]
    confidence = round(100 * max(y_proba[i]), 2)
    print(f"Sampel ke-{i+1}: {label} (Confidence: {confidence}%)")

# ==============================================
# 🌳 Visualisasi Tree
# ==============================================
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=X.columns, class_names=le_target.classes_, filled=True)
plt.title("Decision Tree for Diabetes Diagnosis", fontsize=16)
plt.show()

# ==============================================
# 📜 Ekstrak Aturan
# ==============================================
print("\n📚 Aturan dari Decision Tree:")
rules = export_text(model, feature_names=list(X.columns))
print(rules)

# ==============================================
# 💾 Simpan Model dan Encoder
# ==============================================
joblib.dump(model, 'diabetes_model.pkl')
joblib.dump(le_target, 'label_encoder.pkl')
joblib.dump(le_dict, 'feature_encoders.pkl')
print("\n✅ Model dan semua encoder berhasil disimpan!")
```

```python
# ==============================================
# 📦 Import Library
# ==============================================
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import joblib

# ==============================================
# 📂 Load Dataset
# ==============================================
df = pd.read_csv('diabetes_cleaned_final_with_sehat.csv')
print("📊 Jumlah baris data:", len(df))
print(df.head())

# ==============================================
# 🎯 Pisahkan fitur dan target
# ==============================================
X = df.drop(columns=['Target'])
y = df['Target']

# ==============================================
# 🔠 Encode fitur kategorikal
# ==============================================
le_dict = {}
for col in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    le_dict[col] = le

# ==============================================
# 🔤 Encode label target
# ==============================================
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# ==============================================
# ✂️ Split Data
# ==============================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

# ==============================================
# 🤖 Inisialisasi Model
# ==============================================
model = DecisionTreeClassifier(random_state=42)

# ==============================================
# 🔁 Cross Validation
# ==============================================
cv_scores = cross_val_score(model, X_train, y_train, cv=10)
print("\n📊 Akurasi per Fold:", cv_scores)
print("📈 Rata-rata Akurasi:", np.mean(cv_scores))

# ==============================================
# 🧠 Train Final Model
# ==============================================
model.fit(X_train, y_train)

# ==============================================
# 📉 Evaluasi Model
# ==============================================
y_pred = model.predict(X_test)
print("\n🎯 Akurasi Test Set:", accuracy_score(y_test, y_pred))
print("\n📑 Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=le_target.classes_))

# ==============================================
# 🌳 Visualisasi Decision Tree
# ==============================================
plt.figure(figsize=(10, 5))
plot_tree(model, feature_names=X.columns, class_names=le_target.classes_, filled=True)
plt.title("Decision Tree - Diagnosa Diabetes", fontsize=16)
plt.show()

# ==============================================
# 📜 Ekstrak Aturan
# ==============================================
print("\n📚 Aturan dari Decision Tree:")
print(export_text(model, feature_names=list(X.columns)))

# ==============================================
# 💾 Simpan Model dan Encoder
# ==============================================
joblib.dump(model, 'diabetes_model_v2.pkl')
joblib.dump(le_target, 'label_encoder_v2.pkl')
joblib.dump(le_dict, 'feature_encoders_v2.pkl')

print("\n✅ Model dan semua encoder berhasil disimpan!")
```
